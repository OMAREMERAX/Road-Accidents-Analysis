```python
df.replace([-1, " "], np.nan, inplace=True)
```

```python
df.dropna(subset=["jour", "mois", "an"], inplace=True)
```

```python
numeric_cols = ["jour", "mois", "an"]
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python
categorical_cols = [col for col in df.columns if col not in numeric_cols]
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
df.replace([-1, " "], np.nan, inplace=True)
```

```python
df.dropna(subset=["jour", "mois", "an"], inplace=True)
```

```python
numeric_cols = ["jour", "mois", "an"]
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python
categorical_cols = [col for col in df.columns if col not in numeric_cols]
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
df["date"] = pd.to_datetime(
    df["an"].astype(int).astype(str) + "-" +
    df["mois"].astype(int).astype(str) + "-" +
    df["jour"].astype(int).astype(str),
    errors="coerce"
)

df["hrmn"] = pd.to_datetime(df["hrmn"], format="%H%M", errors="coerce").dt.time
```

```python
severity_map = {1: "Indemne", 2: "Blessé léger", 3: "Blessé hospitalisé", 4: "Tué"}
df["grav"] = df["grav"].map(severity_map)

weather_map = {1: "Normal", 2: "Pluie légère", 3: "Pluie forte", 4: "Neige", 5: "Brouillard"}
df["atm"] = df["atm"].map(weather_map)
```

```python
df.to_csv("data/accidents_cleaned.csv", index=False)
```
